```python
#Indexing in pandas
#iloc and lock
#iloc- integer-based indexing
#loc-label based indexing   #these are more powerful for indexing as we can select random rows also
```

```python
#Normal indexing is limited
#loc["row","column"]
```

```python
#iloc = POSITION-BASED Indexing
data={
    "name":['A','B','C','D','E'],
    "number":[1,2,3,4,5,]
}
df=pd.DataFrame(data,index=["id_1","id_2","id_3","id_4","id_5"])
display(df)
```

```python
df.iloc[2,0] # when i use iloc then it use positioning indexing that means i am specfically saying to return the value at 2 row and 0 columns
```

```python
df.iloc[3,1]
```

```python
df.loc["id_3","number"]  # when i use loc then it considers row and column label instead of index position
```

```python
df.loc[df["number"] > 2, ["name", "number"]]
```

```python
#Manipulating  the index in label based indexing labels can be changed,rename or replaced but in position based indexing we cant change index because it is number and fixed
```

```python
data={
    "name":['A','B','C','D','E'],
    "number":[1,2,3,4,5,]
}
df=pd.DataFrame(data,index=["row_1","row_2","row_3","row_4","row_5"])
display(df)
df.set_index("name",inplace=True)      ##âœ” inplace=True â†’ change the original DataFrame
                                      ##âœ” inplace=False â†’ make a copy, original stays same
display(df)
```

```python
##Conditional selection
data={
    "name":['A','B','C','D','E'],
    "number":[1,2,3,4,5,]
}
df=pd.DataFrame(data,index=["row_1","row_2","row_3","row_4","row_5"])
display(df)
```

```python
df.name=="A"
```

```python
df.loc[df.name=="A"] # df.loc will select the specifc row with the given condition
```

```python
data={
    "Subject":["Hindi","English","Maths","Science","SST"],
    "Marks":[35,36,29,40,21]
}
df=pd.DataFrame(data)
print(df)
```

```python
df.loc[(df.Subject=="Hindi") & (df.Marks>20)]  #iloc will not work here because iloc dont understand label names and consitions
```

```python
data={
    "name":["A","B","C","D","E","F"],
    "marks":[30,40,10,50,60,45],
    "Subject":["Hindi","English","Hindi","Hindi","English","Hindi"]
}
df=pd.DataFrame(data)

df.loc[(df.Subject=="Hindi") & (df.marks>10)]  # and operator - both conditions should satisfy
```

```python
df.loc[(df.Subject=="Hindi")| (df.marks>10)]   # or operator any one condition satify then also return
```

```python
# isin can also be used in this
df.loc[df.Subject.isnull()]
```

```python
df.loc[df.Subject.notnull()]  # return all the values forom dataframe which are not null
```

```python
# Assigning data
df.Subject="Hindi"
display(df)
```
