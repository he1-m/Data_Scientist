```python
# Map- map() works only on a single column (Series).

#It takes each value in the column and applies a function or replacement to it.
```

```python
mean=df.marks.mean()
display(mean)
```

```python
df["Mean_of_marks"]=df["marks"].map(lambda x:x-mean)
display(df)
```

```python
#map works on one value at once while apply workso n multiple values at once
import pandas as pd

df = pd.DataFrame({
    "name": ["Hemant", "Raju", "Mohit"],
    "marks": [50, 80, 90],
    "age": [20, 22, 23]
})

print(df)
```

```python
df["new_marks"]=df["marks"].map(lambda x:x+5)
print(df)  # map takes one value at a time and return one value and dont consider other column
```

```python
df["total_score"]=df.apply(lambda row: row["marks"]+row["age"],axis=1)   # the reason for using row instead of df is that we are sending row to a function not entire dataframe
print(df)  # we are writing axis=1 because we are applying function row wise not column wise
```
